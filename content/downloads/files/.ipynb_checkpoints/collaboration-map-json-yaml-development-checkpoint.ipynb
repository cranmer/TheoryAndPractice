```python
dict_in = yaml.load(document)
```

```python
def expand_dict(dict_in):
    dict_out = {}

    ditem_defaults = {'type':'ditem','links':[], 'description':''}#, 'slug':'slug','date':'2020-01-01 00:00:00'}
    dict_out['ditems'] = []
    n=0
    for i in dict_in['ditems']:
        n+=1
        for key in ditem_defaults.keys() - i.keys():
            i[key]=ditem_defaults[key]
        i['ditem']=n
        dict_out['ditems'].append(i)

    collab_defaults = {'type':'collaboration','description':'','group':1}#,'slug':'slug','date':'2020-01-01 00:00:00'}
    dict_out['collaborations'] = []
    n=0
    for i in dict_in['collaborations']:
        n+=1
        for key in collab_defaults.keys() - i.keys():
            i[key]=collab_defaults[key]
        i['count']=n
        dict_out['collaborations'].append(i)

    proj_defaults = {'type':'project','description':''}#, 'slug':'slug','group':1,'date':'2020-01-01 00:00:00'}
    dict_out['projects'] = []
    n=0
    for i in dict_in['projects']:
        n+=1
        for key in proj_defaults.keys() - i.keys():
            i[key]=proj_defaults[key]
        i['count']=n
        dict_out['projects'].append(i)
        
    return dict_out
        
```

```python
def reduce_dict(dict_in):
    dict_out = {}

    to_remove = set(['type','slug','date','ditem','count','group'])
    dict_out['ditems'] = dict_in['ditems']
    for i in dict_out['ditems']:
        for key in to_remove:
            i.pop(key, None)

    dict_out['collaborations'] = dict_in['collaborations']
    for i in dict_out['collaborations']:
        for key in to_remove:
            i.pop(key, None)

    dict_out['projects'] = dict_in['projects']
    for i in dict_out['projects']:
        for key in to_remove:
            i.pop(key, None)

    return dict_out
        
```

```python
def check_links(dict_in):
    all_links = set()
    for i in dict_in['ditems']:
        all_links.update(i['links'])
    all_projects = set()
    for i in dict_in['projects']:
        all_projects.update([i['name']])
    all_collaborations = set()
    for i in dict_in['collaborations']:
        all_collaborations.update([i['name']])
    print('links exist to the following missing items')
    for i in all_links.difference(all_projects.union(all_collaborations)):
        print('\t',i)
```
